```python
import pandas as pd

# Assume 'purchase_df' is your purchases table and 'sales_df' is your sales table

# 1. Get total purchases and average purchase price for each brand
total_purchases = purchases.groupby('Brand').agg(
    OrderedQuantity=('Quantity', 'sum'), # Renamed for clarity
    PurchasePrice=('PurchasePrice', 'first')
).reset_index()

# 2. Get total sales quantity and sales dollars for each brand
total_sales = sales.groupby('Brand').agg(
    SalesQuantity=('SalesQuantity', 'sum'),
    SalesDollars=('SalesDollars', 'sum')
).reset_index()

# 3. Merge the two tables on the 'Brand' column
summary_df = pd.merge(total_purchases, total_sales, on='Brand', how='left')

# 4. Fill any missing sales (for items never sold) with 0
summary_df['SalesQuantity'] = summary_df['SalesQuantity'].fillna(0)
summary_df['SalesDollars'] = summary_df['SalesDollars'].fillna(0)

# 5. Calculate the remaining quantity
summary_df['RemainingQuantity'] = summary_df['OrderedQuantity'] - summary_df['SalesQuantity']

# 6. Calculate the profit
cost_of_goods_sold = summary_df['SalesQuantity'] * summary_df['PurchasePrice']
summary_df['Profit'] = summary_df['SalesDollars'] - cost_of_goods_sold

# Display the final result with all requested columns
print(summary_df[[
    'Brand',
    'OrderedQuantity',
    'SalesQuantity',
    'RemainingQuantity',
    'Profit'
]])
```

* The purchases table stores detailed purchase information such as purchase date, product/brand bought from vendors, amount paid (in dollars), and quantity purchased.

* The purchase_prices table provides the actual price of each product by vendor and brand, which is unique in this table. The purchase price column in the purchases table is derived from here.

* The vendor_invoice table consolidates data from the purchases table, showing total quantities and dollar values, and also includes an additional freight column. Each record is unique by vendor and purchase order (PO) number.

* The sales table contains details of sales transactions, including the brands bought from vendors, quantity sold, selling price, and revenue generated.

```python
vendor_sales_summary = pd.read_sql_query("""
WITH FreightSummary AS (
    SELECT VendorNumber, VendorName, SUM(Freight) AS FreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber, VendorName
),

PurchaseSummary AS (
    SELECT 
        p.VendorNumber, 
        p.VendorName, 
        p.Brand, 
        MAX(pp.Description) AS Description,   -- ✅ Added Description
        MAX(p.PurchasePrice) AS PurchasePrice,
        MAX(pp.Volume) AS Volume,
        MAX(pp.Price) AS ActualPrice,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand
),

SalesSummary AS (
    SELECT 
        VendorNo,
        VendorName,
        Brand,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(ExciseTax) AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, VendorName, Brand
)

SELECT 
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,              -- ✅ Now works
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss
    ON ps.VendorNumber = ss.VendorNo
   AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC
""", engine)
```

# This query produces a vendor-wise sales and purchase summary, which is useful for:

# Performance Optimization

# Since the query applies multiple joins and aggregations on large tables like sales and purchases, pre-aggregating reduces computational overhead.

# Storing summarized results eliminates the need to repeatedly execute costly operations.

# Facilitates analysis of sales, purchases, and pricing trends across vendors and brands.

# Provides long-term benefits by enabling faster dashboards and reporting.

# Dashboards can directly pull data from vendor_sales_summary, avoiding the need to run heavy queries every time.

```python
from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(text("""
        CREATE TABLE vendor_sales_summary (
            VendorNumber INT,
            VendorName VARCHAR(100),
            Brand INT,
            Description VARCHAR(100),
            PurchasePrice DECIMAL(10,2),
            ActualPrice DECIMAL(10,2),
            Volume DECIMAL(10,2),
            TotalPurchaseQuantity INT,
            TotalPurchaseDollars DECIMAL(15,2),
            TotalSalesQuantity INT,
            TotalSalesDollars DECIMAL(15,2),
            TotalSalesPrice DECIMAL(15,2),
            TotalExciseTax DECIMAL(15,2),
            FreightCost DECIMAL(15,2),
            GrossProfit DECIMAL(15,2),
            ProfitMargin DECIMAL(15,2),
            StockTurnover DECIMAL(15,2),
            SalesToPurchaseRatio DECIMAL(15,2),
            PRIMARY KEY (VendorNumber, Brand)
        );
    """))
```
